# **Single Variant**


```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score,mean_squared_error
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv("/content/data.csv")
df.dropna()

y = df.price.values
x = df.sqft_living.values

X_train, X_val, Y_train, Y_val = train_test_split(x,y,test_size=0.3,random_state=0)

X_train= X_train.reshape(-1, 1)
X_val = X_val.reshape(-1, 1)

lr = LinearRegression()
lr.fit(X_train, Y_train)

lr = LinearRegression()
lr.fit(X_train,Y_train)

y_train_pred = lr.predict(X_train)
y_train_pred

y_val_pred = lr.predict(X_val)
y_val_pred


train_error_score = mean_squared_error(Y_train, y_train_pred) 
print("Mean squared Error - Training : ", train_error_score) 
print("Root mean squared error - Training : ",math.sqrt(train_error_score))

val_error_score = mean_squared_error(Y_val, y_val_pred) 
print("\nMean squared Error - Test : ", val_error_score)
print("Root mean squared Error - Test: ",math.sqrt(val_error_score))

print()
plt.scatter(x,y,color='green')
plt.plot(X_val,y_val_pred,color='blue')
plt.xlabel ('sqft_living')
plt.ylabel ('Price')
plt.title('Linear regression plot for sqft_living & Price')
plt.plot
```

## **Multivariant Regression**


```python
import pandas as pd
import math 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, mean_squared_error
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
from sklearn import metrics

df = pd.read_csv("/content/heart_failure_clinical_records_dataset.csv") 
df = df.dropna()

y = df.DEATH_EVENT
x = df.drop("DEATH_EVENT", axis=1)

X_train, X_val, Y_train, Y_val = train_test_split(x,y,test_size=0.3,random_state=0)

lr = LinearRegression()
lr.fit(X_train,Y_train)

y_train_pred = lr.predict(X_train)
y_train_pred

y_val_pred = lr.predict(X_val)
y_val_pred


train_error_score = mean_squared_error(Y_train, y_train_pred) 
print("Mean squared Error - Training : ", train_error_score) 
print("Root mean squared error - Training : ",math.sqrt(train_error_score))

val_error_score = mean_squared_error(Y_val, y_val_pred) 
print("\nMean squared Error - Test : ", val_error_score)
print("Root mean squared Error - Test: ",math.sqrt(val_error_score))
```
